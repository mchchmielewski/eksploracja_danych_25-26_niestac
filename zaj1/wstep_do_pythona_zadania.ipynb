ZADANIE 1: Analiza danych sprzedażowych
---------------------------------------
Dane: Miesięczna sprzedaż produktu w PLN

sprzedaz = [12500, 13200, 15800, 14200, 16500, 18200, 17800, 19200,
            20500, 21300, 23500, 25200]

a) Oblicz podstawowe statystyki (średnia, mediana, min, max)

b) Znajdź miesiące ze sprzedażą powyżej średniej

c) Oblicz procentowy wzrost sprzedaży (grudzień vs styczeń)

d) Stwórz wykres liniowy pokazujący trend

e) Dodaj linię średniej na wykresie


````python
sprzedaz = [12500, 13200, 15800, 14200, 16500, 18200, 17800, 19200, 20500, 21300, 23500, 25200]
sprzedaz_np = np.array(sprzedaz)
print("a)")
srednia_sprzedaz = np.mean(sprzedaz_np)
print(f"Średnia: {srednia_sprzedaz:.2f}")
print(f"Mediana: {np.median(sprzedaz_np):.2f}")
print(f"Min: {np.min(sprzedaz_np)}")
print(f"Max: {np.max(sprzedaz_np)}")

print("b)")
for i, s in enumerate(sprzedaz):
  if s > srednia_sprzedaz:
    print(i+1)

print("c)")
wzrost = (sprzedaz[11] - sprzedaz[0])/sprzedaz[0] * 100
print(f"{wzrost}%")

print("d,e)")
x = ["styczeń","luty","marzec","kwiecień","maj","czerwiec","lipiec","sierpień","wrzesień","październik","listopad","grudzień"]
plt.plot(x, sprzedaz_np, 'b-', linewidth=2, label='Sprzedaz')
plt.axhline(srednia_sprzedaz)
plt.xlabel('Miesiąc')
plt.ylabel('Sprzedaż')
plt.title('Miesięczna sprzedaż')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

````

ZADANIE 2: Słowniki - Analiza portfolio
---------------------------------------
Stwórz słownik opisujący portfolio inwestycyjne:
```
portfolio = {
    'AAPL': {'liczba': 10, 'cena_zakupu': 150, 'cena_aktualna': 175},
    'GOOGL': {'liczba': 5, 'cena_zakupu': 2800, 'cena_aktualna': 2950},
    'MSFT': {'liczba': 15, 'cena_zakupu': 300, 'cena_aktualna': 320}
}
```

a) Oblicz całkowitą wartość początkową portfolio

b) Oblicz aktualną wartość portfolio

c) Oblicz zysk/stratę dla każdej akcji

d) Które akcje przynoszą największy zwrot (w %)?


````python
portfolio = {
    'AAPL': {'liczba': 10, 'cena_zakupu': 150, 'cena_aktualna': 175},
    'GOOGL': {'liczba': 5, 'cena_zakupu': 2800, 'cena_aktualna': 2950},
    'MSFT': {'liczba': 15, 'cena_zakupu': 300, 'cena_aktualna': 320}
}
print("a)")
calkowita_wartosc_poczatkowa = 0;
for firma in portfolio:
  calkowita_wartosc_poczatkowa += portfolio[firma]['cena_zakupu']
print(calkowita_wartosc_poczatkowa)

print("b)")
calkowita_wartosc_aktualna = 0;
for firma in portfolio:
  calkowita_wartosc_aktualna += portfolio[firma]['cena_aktualna']
print(calkowita_wartosc_aktualna)

print("c)")
zysk_strata = {}
for firma in portfolio:
  zysk_strata[firma] = portfolio[firma]['cena_aktualna'] - portfolio[firma]['cena_zakupu']
  print(f"{firma}: {zysk_strata[firma]}")

print("d)")
zwrot = {}
max_zwrot = 0
firma_zwrot = ""
for firma in portfolio:
  zwrot[firma] = zysk_strata[firma]/portfolio[firma]['cena_zakupu']*100
  if zwrot[firma] > max_zwrot:
    max_zwrot = zwrot[firma]
    firma_zwrot = firma
print(f"{firma_zwrot}: {max_zwrot:.2f}%")

````

ZADANIE 3: Wizualizacja danych klimatycznych
--------------------------------------------
temp_2022 = [2, 4, 8, 13, 18, 21, 24, 23, 19, 14, 8, 3]


temp_2023 = [3, 5, 9, 14, 19, 22, 25, 24, 20, 15, 9, 4]

miesiace = ['Sty', 'Lut', 'Mar', 'Kwi', 'Maj', 'Cze',
            'Lip', 'Sie', 'Wrz', 'Paź', 'Lis', 'Gru']

a) Stwórz wykres liniowy porównujący temperatury w obu latach

b) Zaznacz najcieplejszy i najzimniejszy miesiąc każdego roku

c) Oblicz średnią różnicę temperatur między latami

d) Dodaj odpowiednie etykiety, legendę i tytuł

````python
temp_2022 = [2, 4, 8, 13, 18, 21, 24, 23, 19, 14, 8, 3]
temp_2023 = [3, 5, 9, 14, 19, 22, 25, 24, 20, 15, 9, 4]
miesiace = ['Sty', 'Lut', 'Mar', 'Kwi', 'Maj', 'Cze', 'Lip', 'Sie', 'Wrz', 'Paź', 'Lis', 'Gru']
plt.plot(miesiace, temp_2023, 'r-', linewidth=2, label='temperatury 2023')
plt.plot(miesiace, temp_2022, 'b-', linewidth=2, label='temperatury 2022')
max_2022_index = np.argmax(temp_2022)
max_2023_index = np.argmax(temp_2023)
max_2022_x = miesiace[max_2022_index]
max_2022_y = temp_2022[max_2022_index]
max_2023_x = miesiace[max_2023_index]
max_2023_y = temp_2023[max_2023_index]
plt.annotate(f'max 2022: {max_2022_y:.2f}', xy=(max_2022_x, max_2022_y),
             xytext=(max_2022_x, max_2022_y - 2),
             arrowprops=dict(facecolor='blue', shrink=0.005))
plt.annotate(f'max 2023: {max_2023_y:.2f}', xy=(max_2023_x, max_2023_y),
             xytext=(max_2023_x, max_2023_y + 0.5),
             arrowprops=dict(facecolor='red', shrink=0.005))
min_2022_index = np.argmin(temp_2022)
min_2023_index = np.argmin(temp_2023)
min_2022_x = miesiace[min_2022_index]
min_2022_y = temp_2022[min_2022_index]
min_2023_x = miesiace[min_2023_index]
min_2023_y = temp_2023[min_2023_index]
plt.annotate(f'min 2022: {min_2022_y:.2f}', xy=(min_2022_x, min_2022_y),
             xytext=(min_2022_x, min_2022_y - 0.5),
             arrowprops=dict(facecolor='blue', shrink=0.005))
plt.annotate(f'min 2023: {min_2023_y:.2f}', xy=(min_2023_x, min_2023_y),
             xytext=(min_2023_x, min_2023_y + 2),
             arrowprops=dict(facecolor='red', shrink=0.005))
plt.xlabel('Miesiące')
plt.ylabel('Temperatury')
plt.title('Temperatury w latach 2022/2023')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

roznica = np.array(temp_2023) - np.array(temp_2022)
print(f"\nŚrednia: {np.mean(roznica):.2f}")
````

ZADANIE 4: Symulacja Monte Carlo
---------------------------------
Zasymuluj 1000 rzutów kostką (6-ścienną):

a) Wygeneruj dane używając np.random.randint(1, 7, 1000)

b) Stwórz histogram pokazujący częstość każdej liczby

c) Oblicz teoretyczne i empiryczne prawdopodobieństwo każdej liczby

d) Czy wyniki są zgodne z teorią? (każda liczba ~16.67%)
